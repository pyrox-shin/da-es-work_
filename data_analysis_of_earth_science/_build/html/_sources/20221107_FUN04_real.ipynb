# Loops 迴圈
迴圈 (loop) 的作用是讓某幾行指令在條件符合的情況下一直重覆執行

In Python, loops can be programmed in a number of different ways. The most common is the for loop, which is used together with iterable objects, such as lists. The basic syntax is:

## `for` loops: 
通常是很明確知道要跑「幾次」的這幾行指令。 

### 單層 for-loop:
for 和 in 是 Python 的關鍵字，in的後面則可接一個序列 (Sequence)，迴圈會依序從序列 (例如: 一連串清單資料，一組數字或一個名單) 裡取得元素，並將元素指派給前面自訂的變數，然後執行迴圈裡的內容

```python
for x in ['apple','orange','banana']:
    print("patty likes to eat ",x) 

print("Patty's favorite fruit listed above"  )  
```

###  巢狀for-loop
巢狀迴圈的特性是在迴圈裡面又包覆著其他的迴圈。
外層迴圈每動作一次，內層迴圈就會把整個迴圈執行一遍，執行完畢後才跳回到外層迴圈。

```python
for person in ['patty', 'CM','Alex','Tong']:
    for fruit in ['apple','orange','banana']:
        print(person ,  "likes to eat ",fruit) 
    print(person," favorite fruit listed above!"  ) 
print("We all like to eat fruit!!! ")   
```

```python
for x in np.arange(4):
    print('x=',x)
    for y in [4,5,6]:
        print('we are here y=',y)
        print(x*y)
    print() #猜看看這行在幹嘛？   
```

## `while` loops:
while 是一種無限次數的循環。概念有點像紅綠燈。

如果條件判斷是True，就是綠燈，就做包在裡頭的指令。
如果回來還是綠燈，就再繞一圈，無限循環；
如果條件判斷是False，就是紅燈，就立刻跳出圈圈，不繞了。

```python
i = 0
while i < 5:
    print(i)
    
    i = i + 1
    
print("done")
```
