## * np.random.randn(size) 依照常態分佈趨勢產生隨機數
Return a sample (or samples) from the “standard normal” distribution.

```python
B=np.random.randn(1000)
Bmean = B.mean()
Bstd = B.std()
Bmax = B.max()
Bmin = B.min()
fig, ax = plt.subplots(1,1,figsize=(10,6))
ax.hist(B,bins = 10,density=True,color='c')
ax.set_title('Histogram')

x =  np.linspace(Bmin, Bmax, 100)
mu = Bmean
sigma  = Bstd
p = 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - 0.5*((x - mu)/sigma)**2)
ax.plot(x,p)
ax.set_xlabel('x')
ax.set_ylabel('Probability density function')
```

## np.loadtxt
### 讀檔 way#1
* https://numpy.org/doc/stable/reference/generated/numpy.loadtxt.html

You can create new `numpy arrays` by importing data from files, such as text files. You can import these data using the `loadtxt()` function from numpy, which you imported as np.

For both .txt and .csv files, you need to specify a value for the parameter called fname for the file name (e.g. `np.loadtxt(fname = "filename.txt")`). Be sure to update the path for the file to your home directory.

## 語法- 
### numpy.loadtxt(fname, comments='#', skiprows=0, usecols=None, unpack=False,delimiter=None)

## Options
* comments : str or sequence of str, optional
The characters or list of characters used to indicate the start of a comment. None implies no comments. For backwards compatibility, byte strings will be decoded as ‘latin1’. The default is ‘#’.
* read the columns of the interests. 
 usecols : int or sequence, optional Which columns to read, with 0 being the first. For example, usecols = (0,1,2) will extract the 1st, 2nd and 3thcolumns. The default, None, results in all columns being read.

* unpack : bool, optional If True, the returned array is transposed, so that arguments may be unpacked using x, y, z = loadtxt(...). When used with a structured data-type, arrays are returned for each field. Default is False.

## 變數
* fname : 檔案名稱
* comments : 註解符號，預設為'#'。
* skiprows : 跳過前面n行（預設為0）
* usecols : 要讀入的直行數（預設為所有的直行都讀入）
* unpack : 是否要將一個直行讀取成一個1D array，預設為關閉(False)，If set to 'True", x, y, z = loadtxt(...)
